```python
from matplotlib import pyplot as plt
from PIL import Image
import numpy as np
import PIL.ImageOps
import csv
import zipfile
import os
import cv2
import pandas as pd
```

```python
# dir_path = "/content/drive/MyDrive/malaria_project/landcover_cropped_with_confi_2010_2018_size10andGreater/"
dir_path = "landcover_1992onwards_min5examined_cropped_17x17"

# output_folder = "/content/drive/MyDrive/malaria_project/landcover_one_hot_cropped_with_confi_2010_2018_size10andGreater/"
output_folder = "landcover_1992onwards_min5examined_cropped_17x17_oneHotEncoded"
```

```python
for i, file in enumerate(os.listdir(dir_path)):
    print(i, "/", len(os.listdir(dir_path)))
    if file == ".DS_Store" or file == ".ipynb_checkpoints":
        continue
    # if i < 2630:
    #   continue
    img = Image.open(dir_path + "/" + file)

    #convert to onehot image with 38 channels, one for each land cover class
    # print(img)
    arr = np.array(img)
    # print(arr)

    '''
	stuff to search:
	one hot encode input image
	categorical matrix into one-hot encoded images
	rgb segmentation masks to classes tensorflow
	CategoryEncoding keras make 3d encoding
	'''

    value_list = [0, 10, 11, 12, 20, 30, 40, 50, 60, 61, 62,
                  70, 71, 72, 80, 81, 82, 90, 100, 110,
                  120, 121, 122, 130, 140, 150, 151, 152, 153, 160, 170, 180,
                  190, 200, 201, 202, 210, 220]
    # print(len(value_list))

    one_hot_tensor = np.zeros((17, 17, 38))

    a = np.where(arr == 10)
    # print(a)
    for i, val in enumerate(value_list):
        indx = np.where(arr == val)
        one_hot_tensor[:, :, i] = arr
        # print( one_hot_tensor[:,:,i])
        one_hot_tensor[:, :, i][one_hot_tensor[:, :, i] == val] = 1
        one_hot_tensor[:, :, i][one_hot_tensor[:, :, i] != 1] = 0
        # print( one_hot_tensor[:,:,i])



    # print(one_hot_tensor.shape)

    # print(one_hot_tensor[:,:,5])
    # print(one_hot_tensor[:,:,6])

    #save image to folder
    file_wo = file.split('.')[0]
    np.save(output_folder + "/" + file_wo + '.npy', one_hot_tensor)
```
